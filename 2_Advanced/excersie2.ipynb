#################### INVESTIGATE TRENDING SKILLS  ##############################

```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
df_DA
```

```python
# extracting month

df_DA['job_posted_month'] = df_DA['job_posted_date'].dt.month
df_DA
```

```python
# EXPLODING THE LIST OF SKILLS
df_DA_exploded = df_DA.explode('job_skills')
df_DA_exploded
```

```python
# PIVOTING THE DATAFRAME TO GET SKILL COUNTS

df_DA_pivot = df_DA_exploded.pivot_table(index='job_posted_month', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
df_DA_pivot

# SORTING THE DATAFRAME TO GET TOP 10 SKILLS
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_pivot = df_DA_pivot.drop('Total')
```
